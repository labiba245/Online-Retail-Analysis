```python
new_df["RevenueGenerated"] = new_df["Quantity"] * new_df["UnitPrice"]
```

# Repeat Customers

```python
invoice = (
    new_df.groupby(by=["InvoiceNo", "InvoiceDate"])
    .agg({"RevenueGenerated": sum, "CustomerID": max, "Country": max,})
    .reset_index()
)
invoice
```

```python
# total monthly revenue
monthly_revenue = (
    new_df.set_index("InvoiceDate")["RevenueGenerated"].resample("M").sum()
)

# total monthly revenue from repearn customers
monthly_revenue_repeat_customers = (
    invoice.set_index("InvoiceDate")
    .groupby([pd.Grouper(freq="M"), "CustomerID"])
    .filter(lambda x: len(x) > 1)
    .resample("M")
    .sum()["RevenueGenerated"]
)

# percentage of the revenue from repeat customers to the total monthly revenue
percent = (monthly_revenue_repeat_customers / monthly_revenue * 100.0).round(2)
percent
```

```python
monthly_plot = pd.DataFrame(monthly_revenue.values).plot(figsize=(15, 8))

pd.DataFrame(monthly_revenue_repeat_customers.values).plot(
    ax=monthly_plot, grid=True,
)

monthly_plot.set_xlabel("Date")
monthly_plot.set_ylabel("Sales")
monthly_plot.set_title("Total Revenue vs. Revenue from Repeat Customers")

monthly_plot.legend(["Total Revenue", "Revenue from repeat customer"])

repeat_plot = monthly_plot.twinx()

pd.DataFrame(percent.values).plot(ax=repeat_plot, kind="bar", alpha=0.4)

repeat_plot.set_ylabel("percentage (%)")
repeat_plot.legend(["Repeat Revenue Percentage"])

monthly_plot.set_xticklabels([x.strftime("%Y-%m") for x in percent.index])

plt.show()
```

# Item Analysis

```python
# top 10 items sold
top_ten_items_sold = pd.DataFrame(
    new_df.groupby("StockCode")["Quantity"].sum().sort_values(ascending=False).head(10)
)
top_ten_items_sold = top_ten_items_sold.reset_index()
top_ten_items_sold
```

```python
top_ten_items_sold_list = top_ten_items_sold["StockCode"].tolist()
```

```python
top_sold_items = new_df.loc[new_df["StockCode"].isin(top_ten_items_sold_list)]
top_sold_items
```

```python
# top sold items sales per month
sales_per_month = (
    top_sold_items.set_index("InvoiceDate")
    .groupby([pd.Grouper(freq="M"), "StockCode"])["Quantity"]
    .sum()
)
sales_per_month
```

```python
# create a pivot table to summarize the above data
summary = sales_per_month.reset_index().pivot("InvoiceDate", "StockCode").fillna(0)
summary = summary.reset_index()
summary = summary.set_index("InvoiceDate")
summary.columns = summary.columns.droplevel(0)
summary
```

```python
# item trend

trend = pd.DataFrame(summary.values).plot(figsize=(15, 8))

trend.set_ylabel("Quantity")
trend.set_xlabel("Month")
trend.set_title("Top 10 sold items sales trend")

trend.legend(summary.columns)

plt.xticks(
    range(len(summary.index)),
    [x.strftime("%Y-%m") for x in summary.index],
    rotation=45,
)

plt.show()
```
